```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import csv
import requests
```

```python
f1=pd.read_csv('airdatasetnew.csv',encoding='unicode_escape')
f1.dtypes
```

```python
f1['date'] = pd.to_datetime(f1['date'], errors='coerce')
f1['year']=f1['date'].dt.year
f1.head()
```

```python
f1.drop(['sn','spm','rspm','location'],axis=1,inplace=True)

f1.head()
```

```python
option=['Delhi','Haryana','Maharashtra','Kerala','Tamil Nadu','West Bengal','Gujarat']
f1=f1.loc[(f1['state'].isin(option)) & (f1['date'].dt.year>=2000)]

f1.head()
f1.fillna(0)
f1=f1.replace(np.nan,0)
```

```python
dd=f1.groupby(['year','state']).agg('mean')
dd
```

```python
dataset2=pd.read_csv('pune1.csv')
dataset2.head()
```

```python
#Finding missing values in the data set 
dataset2=dataset2.drop(['pm10','o3'],axis=1)
dataset2.head()
dataset2.fillna(dataset2.mean())
dataset2=dataset2.replace(np.nan,0)

dataset2['date'] = pd.to_datetime(dataset2['date'], errors='coerce')
dataset2['year']=dataset2['date'].dt.year
dataset2=dataset2.loc[(dataset2['date'].dt.year>2015)]
dataset2.head()
```

```python
dataset2=dataset2.fillna(dataset2.mean())
dataset2=dataset2.replace(np.nan,0)
dataset2 = dataset2.dropna()
```

```python
print(dataset2.head())
```

```python
dataset2.fillna(" ")
```

```python
dataset2.isnull().sum()
```

```python
dataset2.dtypes
```

```python
for col in dataset2.iloc[:, 2:].columns: 
    dataset2[col] = pd.to_numeric(dataset2[col], errors='coerce')
dataset2.dtypes
```

```python
def calculate(co2,no2,pm2_5):
   result=0
   result=((co2/80)+(no2/80)+(pm2_5/75))*100 
   return result
dataset2['AQI']=dataset2.apply(lambda x:calculate(x['co2'],x['no2'],x['pm2_5']),axis=1)
dataset2.head()
dataset2.rename(columns={'co2':'so2'},inplace=True)
```

```python
df1=dataset2.groupby(['year','state']).agg('mean')
df1.head()
```

```python
aa=pd.concat([dd,df1],sort=False,axis=0)
y1=aa.to_csv('aircsv.csv')
```
