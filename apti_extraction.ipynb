```python
import io
import json
import cv2
import numpy as np
import requests
```

```python
img = cv2.imread("Vol13_No1_Are_Lak_T1.jpg")
url = "https://api.ocr.space/parse/image"
_,compressed_img = cv2.imencode(".jpg", img , [1,90])
file_bytes = io.BytesIO(compressed_img)

results = requests.post(url, 
                        files = {"Vol13_No1_Are_Lak_T1.jpg" : file_bytes}, 
                        data = {"apikey" : "c6cd621c4288957", "language" : "eng"})
```

```python
results = results.content.decode()
results = json.loads(results)
parsed_results = results.get("ParsedResults")[0]
text_detected = parsed_results.get("ParsedText")
print(text_detected)
```

```python
li = list(text_detected.split(","))
print(li)
col1 = [i for i, s in enumerate(li) if 'Plant species (vegetation)\r' in s]
col1 = int(col1[0])
col2 = [i for i, s in enumerate(li) if '(Air Pollution Tolrance\r' in s]
col2 = int(col2[0])
print(col1)
print(col2)
```

```python
import csv
from itertools import zip_longest

species = li[col1:col2]
apti = li[col2:]

lis = [species,apti]
columns = ['Plant Species', 'APTI']

with open("APTI_Values.csv", 'w', newline='') as csvfile:
    write = csv.writer(csvfile)
    write.writerow(columns)
    write.writerows(zip_longest(*lis, fillvalue=''))
    
```
