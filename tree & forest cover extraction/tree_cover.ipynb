```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.frienvis.nic.in/Database/Tree-Cover-in-Indian-States-and-UTs_2468.aspx"
page = requests.get(url)
```

```python
print(page.status_code)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
tb = soup.find('table')
```

```python
for row in tb.find_all('td'):
    name = row.find('span')
    print(name)
```

```python
data = []
for row in tb.find_all('td'):
    name = row.find('span')
    if name is not None:
        str = name.get_text()
    print(str)
    data.append(str)
```

```python
dt = ['State/UT', 'Tree_Cover-2001', 'Tree_Cover-2003', 'Tree_Cover-2005', 'Tree_Cover-2009', 'Tree_Cover-2011',
      'Tree_Cover-2013', 'Tree_Cover-2015', 'Tree_Cover-2017', 'Tree_Cover-2019']
print(dt)
```

```python
it = [iter(dt)] * 10
```

```python
tuples = []
for l in zip(*it):
    print(l)
    tuples.append(l)
```

```python
it = [iter(d)] * 9
```

```python
for l in zip(*it):
    print(l)
    tuples.append(l)
```

```python
print(tuples)
```

```python
df = pd.DataFrame(tuples)
print(df)
```

```python
df.head()
```

```python
df1 = df.rename(columns=df.iloc[0])
df1.head()
```

```python
tree = df1.drop(df1.index[0])
tree.head()
```

```python
tree.to_csv('tree_cover.csv', na_rep='Null', float_format='%.2f', index=False, encoding='utf-8')
```
