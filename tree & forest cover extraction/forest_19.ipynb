```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.frienvis.nic.in/Database/Forest-Cover-in-States-UTs-2019_2478.aspx"
page = requests.get(url)
```

```python
print(page.status_code)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
tb = soup.find('table')
```

```python
for row in tb.find_all('td'):
    name = row.find('div')
    print(name)
```

```python
data = []
for row in tb.find_all('td'):
    name = row.find('div')
    if name is not None:
        str = name.get_text()
    print(str)
    data.append(str)
```
