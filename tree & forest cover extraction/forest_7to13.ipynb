```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
```

```python
url = "http://www.frienvis.nic.in/Database/Type-of-Forest-Cover-in-India-2007-13_2249.aspx"
page = requests.get(url)
```

```python
print(page.status_code)
```

```python
print(page.content)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
print(soup.prettify())
```

```python
tb = soup.find('table')
```

```python
for row in tb.find_all('td'):
    name = row.find('span')
    print(name)
```

```python
data = []
for row in tb.find_all('td'):
    name = row.find('span')
    if name is not None:
        str = name.get_text()
    print(str)
    data.append(str)
```
