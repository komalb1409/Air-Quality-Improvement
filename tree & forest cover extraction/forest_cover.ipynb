```python
import pandas as pd
import csv
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
f1 = pd.read_csv('forest_1to13.csv')
f1.head()
```

```python
f2 = pd.read_csv('forest_15to19.csv')
f2.head()
```

```python
f3 = pd.DataFrame(f1)
cols = [0, 1]
f3 = f3[f3.columns[cols]]
```

```python
f3.rename(columns={'State/UT':'State/UT', 'Geographical Area':'Geographical Area_1-13'}, inplace=True)
```

```python
f3.head()
```

```python
f4 = pd.DataFrame(f2)
cols = [1]
f4 = f4[f4.columns[cols]]
```

```python
f4.rename(columns={'Geographical Area':'Geographical Area_15-19'}, inplace=True)
```

```python
f4.head()
```

```python
f5 = pd.concat([f3, f4], axis=1, sort=False)
```

```python
f5.head()
```

```python
f6 = f1.drop(['State/UT', 'Geographical Area'], axis = 1)
f6.head()
```

```python
f7 = pd.concat([f5, f6], axis = 1, sort = False)
f7.head()
```

```python
f8 = f2.drop(['State/UT', 'Geographical Area'], axis = 1)
f8.head()
```

```python
forest_cover = pd.concat([f7, f8], axis = 1, sort = False)
forest_cover.head()
```

```python
forest_cover.to_csv('forest_cover.csv', na_rep='Null', float_format='%.2f', index=False, encoding='utf-8')#area in km2
```
