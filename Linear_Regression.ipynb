```python
import pandas as pd
air = pd.read_csv("india-air-quality-data2.csv",encoding='cp1252')
year = 2001
years = []
while year<=2015:
    years.append(year)
    year = year+2
s = air['state']
s = list(s)
s = list(set(s))
s.remove('Telangana')
s.remove('Uttaranchal')
print(years)
print(s)
```

```python
aqs = []
st_av = []
y_av = []
air['date'] = pd.to_datetime(air['date'])
for st in s:
    for y in years:
        aq = air['AQI'].loc[(air['date'].dt.year == y) & (air['state']==st)]
        aq = list(aq)
        if len(aq) != 0:
            avg_aq = sum(aq)/len(aq)
            aqs.append(avg_aq)
            st_av.append(st)
            y_av.append(str(y))
            print(st,y)
print(aqs)
```

```python
print(st_av)
print(y_av)
print(len(aqs))
```

```python
trees = []
ttc = pd.read_csv("Total Tree Cover.csv")
for i in range(len(st_av)):
    t = ttc[y_av[i]].loc[ttc['States'] == st_av[i]].item()
    trees.append(t)
    print(st_av[i],y_av[i])
print(trees)
```

```python
print(len(trees))
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn import metrics  
import matplotlib.pyplot as plt

a = []
b = []

for i in range(1,51):
    x = np.array(aqs).reshape((-1,1))
    y = np.array(trees)
    #print(x)
    #print(y)

    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state=0)
    #print(x_train)
    #print(x_test)
    #print(y_train)
    #print(y_test)
    transformer = PolynomialFeatures(degree=i, include_bias=False)
    transformer.fit(x_train)
    x_ = transformer.transform(x_train)
    #print(x_)

    model = LinearRegression()
    model.fit(x_,y_train)
    transformer.fit(x_test)
    x_test_ = transformer.transform(x_test)
    y_pred = model.predict(x_test_)
    #print(y_pred)

    ssr = np.sum((y_pred - y_test)**2)
    sst = np.sum((y_test - np.mean(y_test))**2)
    r2_score = 1 - (ssr/sst)
    #print('ssr',ssr)
    #print('sst',sst)
    #print('r2_score',r2_score)
    a.append(i)
    b.append(r2_score)
plt.plot(a,b,color='r')
plt.xlabel('Degree')
plt.ylabel('R^2')
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
from sklearn import metrics  
import matplotlib.pyplot as plt

sizes = [0.5,0.6,0.7,0.8,0.9]
accuracy = []
for size in sizes:
    x = np.array(aqs).reshape((-1,1))
    y = np.array(trees)
    x_train, x_test, y_train, y_test = train_test_split(x,y,train_size = size, random_state=0)
    #print(x_train)
    #print(x_test)
    #print(y_train)
    #print(y_test)
    transformer = PolynomialFeatures(degree=9, include_bias=False)
    transformer.fit(x_train)
    x_ = transformer.transform(x_train)
    #print(x_)

    model = LinearRegression()
    model.fit(x_,y_train)
    transformer.fit(x_test)
    x_test_ = transformer.transform(x_test)
    y_pred = model.predict(x_test_)
    #print(y_pred)
    ssr = np.sum((y_pred - y_test)**2)
    sst = np.sum((y_test - np.mean(y_test))**2)
    r2_score = 1 - (ssr/sst)
    accuracy.append(r2_score)

config = ['50:50','60:40','70:30','80:20','90:10']
fig, ax = plt.subplots()
ax.plot(config,accuracy)
ax.set(xlabel = 'Configurations', ylabel = 'Accuracies')
```
